```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
# Charger les données
data = pd.read_excel('donnees.xlsx')
data['t(min)'] = data['t(min)'].apply(math.ceil)
bar_data = data.melt(id_vars=['Observation', 'V(C/min)', 't(min)', 'T'],
                     value_vars=['Experimental', 'Random Forest Output'],
                     var_name='Type', value_name='Yield')
data=data.dropna()
data
```

```python
# Graphique pour la température
plt.figure(figsize=(14, 8))
sns.barplot(x='T', y='Yield', hue='Type', data=bar_data, ci=None)
plt.xlabel('Temperature (°C)')
plt.ylabel('Hydrogen Yield')
plt.title('Comparison of Experimental and Random Forest Predicted Hydrogen Yield by Temperature')
plt.legend(title='Data Type')
plt.tight_layout()
plt.show()
```

```python
# Graphique pour la vitesse de chauffage
plt.figure(figsize=(14, 8))
sns.barplot(x='V(C/min)', y='Yield', hue='Type', data=bar_data, ci=None)
plt.xlabel('Heating Rate (°C/min)')
plt.ylabel('Hydrogen Yield')
plt.title('Comparison of Experimental and Random Forest Predicted Hydrogen Yield by Heating Rate')
plt.legend(title='Data Type')
plt.tight_layout()
plt.show()
```

```python
# Graphique pour le temps
plt.figure(figsize=(14, 8))
sns.barplot(x='t(min)', y='Yield', hue='Type', data=bar_data, ci=None)

plt.xlabel('Time (min)')
plt.ylabel('Hydrogen Yield')
plt.title('Comparison of Experimental and Random Forest Predicted Hydrogen Yield by Time')
plt.legend(title='Data Type')
plt.tight_layout()
plt.show()
```
